```python
import numpy as np 
A = np.array([[1.44,-0.36,5.52,0.00],[-0.36,10.33,-7.78,0.00],[5.52,-7.78,28.40,9.00],[0.00,0.00,9.00,61.00]])
print(A)
print("----------------------------------------------------------------------------------------")
A_ = np.linalg.inv(A) #inverse the Matrix A
print(A_)
print("----------------------------------------------------------------------------------------")
B = np.array([[0.04],[-2.15],[0],[0.88]])
print(B)

#A multiply A to its inverse ----------- AA^-1X = B*A^-1
print("----------------------------------------------------------------------------------------")
#AXA'
AA_ = np.dot(A,A_)
print(AA_)
print("----------------------------------------------------------------------------------------")
#A'X B
BA_ = np.dot(A_,B)
print("the Unknown Variables Are:")
print(BA_)
print("----------------------------------------------------------------------------------------")
print("proving:")
print(np.dot(A,BA_))
```
